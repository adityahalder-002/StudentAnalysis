```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_scores.csv')
print(df.head())
```

```python
df.describe() #give information of numeric value
```

```python
df.info() #give information of datatypes
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

Gender Distribution

```python
plt.figure(figsize = (5,5))
plt.title("Gender Distribution")
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we have analyzed that no of female is more as compared to male
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
plt.title("Reltionship Between Parent's Education and Students Score")
sns.heatmap(gb,annot = True)
plt.show()
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (4,4))
plt.title("Relationship betwwen Parent's Maritial Status and Student's Score")
sns.heatmap(gb1,annot= True)
plt.show()
```

```python
#from the above fig we conclue that parents marital status have negligible impact on students marks
```

Distribution of Ethnic Group

```python
groupA = df.loc[(df['EthnicGroup']=='group A')].count()
groupB = df.loc[(df['EthnicGroup']=='group B')].count()
groupC = df.loc[(df['EthnicGroup']=='group C')].count()
groupD = df.loc[(df['EthnicGroup']=='group D')].count()
groupE = df.loc[(df['EthnicGroup']=='group E')].count()

plt.title("Distribution of Ethnic Group")
l = ["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
print(mlist)
```
